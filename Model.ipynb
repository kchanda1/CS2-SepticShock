```python
import pandas as pd

df=pd.read_csv('CleanData.csv')
df.head()
```

```python
df.drop(['DBP', 'BUN', 'Glucose','EtCO2', 'BaseExcess', 'HCO3', 'FiO2','pH', 'PaCO2','SaO2', 'AST', 'Alkalinephos', 'Calcium', 'Chloride', 'Creatinine', 'Bilirubin_direct', 'Lactate', 'Magnesium', 'Phosphate', 'Potassium', 'Bilirubin_total','TroponinI','Hct', 'Hgb', 'PTT', 'WBC', 'Fibrinogen', 'Platelets'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df = df.reset_index()
```

```python
df.shape
```

```python
# Model Naive Bayes

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix


# Load data
# df=pd.read_csv("CleanData.csv")


Y=df.pop("Shock")
X=df

# Divinding dataset into training and test
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.4, stratify=Y)
X_train, X_val, Y_train, Y_val=train_test_split(X_train, Y_train, test_size=0.4, stratify=Y_train)

model = GaussianNB()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print("Accuracy Score for Naive Bayes: \n" , accuracy_score(Y_test, Y_pred))
print("Confusion Matrix for Naive Bayes: \n" , confusion_matrix(Y_test, Y_pred))
```
