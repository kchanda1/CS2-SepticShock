```python
import pandas as pd
import plotly.express as px

dataset=pd.read_csv('Dataset1.csv')
dataset.head(10)
```

```python
septic_id=[]
test=1
for row in dataset.itertuples():
    if test !=row.Patient_ID:
        if row.SepsisLabel == 1:
            test=row.Patient_ID
            septic_id.append(row.Patient_ID)
            
number_of_elements_septic = len(septic_id)

print("Total number of unique Patient IDs: ", number_of_elements_septic)
```

```python
full_id=[]

test2=1
for row in dataset.itertuples():
    if test2 !=row.Patient_ID:
        test2=row.Patient_ID
        full_id.append(row.Patient_ID)

    
# print(full_id)

number_of_elements_full = len(full_id)

print("Total number of unique Patient IDs: ", number_of_elements_full)
```

```python
non_septic_id = list(set(full_id) - set(septic_id))

number_of_elements_nonseptic = len(non_septic_id)


print("Total number of unique non-septic Patient IDs: ", number_of_elements_nonseptic)
```

```python
# Data for patients that did not go into septic shock

import pandas as pd

data = pd.read_csv('Dataset1.csv')

for i,value in enumerate(non_septic_id):
    data[data['Patient_ID'] == value].to_csv(r'Patient_ID_'+str(value)+r'.csv',index = False, na_rep = 'N/A')
```
