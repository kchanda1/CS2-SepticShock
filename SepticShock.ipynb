```python
# Data for patients that went into septic shock

import pandas as pd

data = pd.read_csv('Dataset1.csv')

#     data_category_range = data['Category'].unique()
#     data_category_range = data_category_range.tolist()

for i,value in enumerate(septic_id):
#     value.to_csv(f'{SepticShock}.csv', index=False)
    data[data['Patient_ID'] == value].to_csv(r'Patient_ID_'+str(value)+r'.csv',index = False, na_rep = 'N/A')
```

```python
# Data for patients who didnt go into septic shock

import pandas as pd

data = pd.read_csv('Dataset1.csv')

#     data_category_range = data['Category'].unique()
#     data_category_range = data_category_range.tolist()

for i,value in enumerate(septic_id):
#     value.to_csv(f'{SepticShock}.csv', index=False)
    data[data['Patient_ID'] != value].to_csv(r'Patient_ID_'+str(value)+r'.csv',index = False, na_rep = 'N/A')
```
