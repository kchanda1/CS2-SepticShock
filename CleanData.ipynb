```python
import pandas as pd

dataset=pd.read_csv('SepticShockAll.csv')
dataset.head(10)
```

```python
# Removing rows when patient has gone to shock stage
dataset_BeforeShock=dataset[dataset['SepsisLabel'] != 1]
dataset_BeforeShock.head(200)
```

```python
# Average values of parameters by Patient ID 

df=dataset_BeforeShock.groupby(['Patient_ID']).mean()
df.head(10)
```

```python
df.drop(['Hour', 'Unit1', 'Unit2', 'HospAdmTime', 'ICULOS', 'SepsisLabel'], axis = 1, inplace = True)
```

```python
df.drop(df.columns[[0]], axis = 1, inplace = True)
df.head(10)
```

```python
df.insert(0, 'Shock', '1')
df.head()
```

```python
dataset_nss=pd.read_csv('NonSShockAll.csv')

# Average values of parameters by Patient ID 

df_nss=dataset_nss.groupby(['Patient_ID']).mean()
df_nss.drop(['Hour', 'Unit1', 'Unit2', 'HospAdmTime', 'ICULOS', 'SepsisLabel'], axis = 1, inplace = True)
df_nss.drop(df_nss.columns[[0]], axis = 1, inplace = True)
df_nss.insert(0, 'Shock', '0')
df_nss.head()
```

```python
df.shape
```

```python
df_nss.shape
```

```python
# Combining data in one csv for people who got septic shock and people who did not

pd.concat([
    pd.concat([df], axis=1),
    pd.concat([df_nss], axis=1)]).to_csv('CleanData.csv')
```
